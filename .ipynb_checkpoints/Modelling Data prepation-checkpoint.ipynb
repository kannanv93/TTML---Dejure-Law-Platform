```python
ranks = []
test_corpus_facts_vec = np.zeros((len(test_corpus_facts),50))


for doc_id in tqdm(range(len(test_corpus_facts))):
    inferred_vector = fact_model.infer_vector(test_corpus_facts[doc_id].words)
    test_corpus_facts_vec[doc_id,:] = inferred_vector
    sims = fact_model.dv.most_similar([inferred_vector], topn=4)
    rank = [docid for docid, sim in sims]
    ranks.append(rank[1:])
```

```python
ranks
```

```python
df_cmb['Similar_facts_index'] = ranks
```

```python
judgement_model = gensim.models.doc2vec.Doc2Vec(vector_size=50, min_count=2, epochs=20)

judgement_model.build_vocab(test_corpus_judge)

judgement_model.train(test_corpus_judge, total_examples=judgement_model.corpus_count, epochs=judgement_model.epochs)
```

```python
jud_ranks = []
test_corpus_judgement_vec = np.zeros((len(test_corpus_judge),50))


for doc_id in tqdm(range(len(test_corpus_judge))):
    inferred_vector = judgement_model.infer_vector(test_corpus_judge[doc_id].words)
    test_corpus_judgement_vec[doc_id,:] = inferred_vector
    sims = judgement_model.dv.most_similar([inferred_vector], topn=4)
    rank = [docid for docid, sim in sims]
    jud_ranks.append(rank[1:])
```

```python
df_cmb['Similar_judgement_index'] = jud_ranks
```
