```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
import shutil
import re
```

```python
for i in range(1900,2000):
    try:
        shutil.rmtree("/scratch/hrr288/Dejure/kolkata/"+str(i)+"/")
    except:
        continue
```

```python
crime_list = ['crl.o.p','crl','criminal original petition','crime','cr.p.c','crl','criminal']
regex = re.compile("(?=(" + "|".join(map(re.escape, crime_list)) + "))")


return_list=[]
for i in range(2000,2020):
    for filenames in os.walk('./kolkata/'+str(i)):
        for file_list in filenames:
            for file_name in file_list:
                if file_name.endswith((".txt")):
                    return_list.append('/scratch/hrr288/Dejure/kolkata/'+str(i)+"/"+file_name)

return_list.sort()
```

```python
len(return_list)
```

```python
crime_cases = []

for i in tqdm(range(len(return_list))):
    try:
        #print(i)
        text= open(return_list[i],'r').read()
        parsed_html = BeautifulSoup(text)
        text = parsed_html.body.find('div', attrs={'class':'judgments'}).text
        text = ' '.join(text.split()).replace('http://www.judis.nic.in',"").lower()
        #print(text)
        #print(len(re.findall(regex, text)))
        if(len(set(re.findall(regex, text)))>1 ):
            crime_cases.append(return_list[i])
    except:
        continue
```

```python
crime_df = pd.DataFrame(crime_cases,columns=['Path'])
crime_df.to_csv("/scratch/kv942/Dejure/kolkata_2000_20.csv",index=False)
```
